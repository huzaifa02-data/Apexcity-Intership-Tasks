# Task 3

```python
df3 = pd.read_csv("initial-disability-and-hearing-level-service-satisfaction.csv")
df3.head()
```

```python
df3.drop(['FY2013','FY2015','FY2016','FY2017','FY2018'],axis=1)
```

```python
df3['FY2008'].value_counts()
```

```python
df3['FY2014'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
years = ["FY2008", "FY2009", "FY2010", "FY2011", "FY2012", "FY2014"]

for year in years:
    plt.figure(figsize=(10,6))
    sns.barplot(data=df3, x=year, y="ASPECT OF SERVICE SATISFACTION")
    plt.title(f"{year} Service Satisfaction")
    plt.tight_layout()
    plt.show()
```
